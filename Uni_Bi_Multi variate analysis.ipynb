```python
import pandas as pd 
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data, columns= iris.feature_names)
table.head()
```

```python
table['Target']=iris.target
table.head()
```

```python
table.shape
```

```python
setosa = table.loc[table['Target']==0]
setosa
```

```python
virginica = table.loc[table['Target']==1]
virginica
```

```python
versicolor = table.loc[table['Target']==2]
versicolor
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.plot(setosa['sepal width (cm)'],np.zeros_like(setosa['sepal width (cm)']),'s')
plt.plot(virginica['sepal width (cm)'],np.zeros_like(virginica['sepal width (cm)']),'o')
plt.plot(versicolor['sepal width (cm)'], np.zeros_like(versicolor['sepal width (cm)']),'d')
plt.xlabel('sepal width')
```

```python
np.zeros_like(versicolor['sepal length (cm)'])
```

```python
plt.scatter(table['sepal length (cm)'], table['petal length (cm)'], c= table['Target'])
```
